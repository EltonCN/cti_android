```python
import urllib.request
from urllib.error import URLError
import numpy as np
import json
```

```python
url = "http://192.168.2.101:8080/sensors.json"
```

```python
import time

tempo = time.perf_counter()

tempoHist = []

accel = []

ultimoIndice = 0

indice = []

try:
    for i in range(1):
        with urllib.request.urlopen(url) as req:
            data = req.read()

            dataDict = json.loads(data)

            tempoHist.append(time.perf_counter()-tempo)
            tempo = time.perf_counter()

            for j in range(len(dataDict["accel"]["data"])):
                if(dataDict["accel"]["data"][j][0]> ultimoIndice):
                    ultimoIndice = dataDict["accel"]["data"][j][0]
                    indice.append(dataDict["accel"]["data"][j][0])
                    accel.append(dataDict["accel"]["data"][j][1])
except URLError:
    print("Não conseguiu conectar")
```

```python
from matplotlib import pyplot as plt

plt.plot()
plt.show()
```

```python
indiceFormat = []

for i in range(len(indice)-1):
    indiceFormat.append(indice[i+1]-indice[i])

plt.plot(indiceFormat)
plt.show()

np.mean(indiceFormat)
```
